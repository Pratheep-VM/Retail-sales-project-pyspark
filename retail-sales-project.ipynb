```python
dbutils.widgets.dropdown("timeperiod","weekly",["weekly","monthly"])
```

```python
from datetime import date, timedelta, datetime
from pyspark.sql.functions import *

# 1. Get the value selected in the dropdown menu
time_period = dbutils.widgets.get("timeperiod")
print(time_period)

# 2. Get today's date
today = date(2024, 7, 27)

if time_period == 'weekly':
    # 3. Calculate start date for 'Weekly' (Previous Week)
    # Finds the previous week's start day based on today's weekday
    start_date = today - timedelta(days=today.weekday(), weeks=1) - timedelta(days=1)
    
    # 4. Calculate end date (6 days after start date)
    end_date = start_date + timedelta(days=6)
    
else:
    # 5. Calculate logic for 'Monthly' (Previous Month)
    # Get the first day of the CURRENT month
    first = today.replace(day=1)
    
    # Go back one day from the 1st to get the last day of the PREVIOUS month
    end_date = first - timedelta(days=1)
    
    # Go back the number of days in the previous month to get the 1st of PREVIOUS month
    start_date = first - timedelta(days=end_date.day)

print(start_date, end_date)
```
